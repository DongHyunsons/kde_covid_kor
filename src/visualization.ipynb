```python
def filter(data, threshold):
    tmp = data[data[:] >= threshold]
    tmp[tmp[:].isnull() == False] = 0
    tmp[tmp[:].isnull()] = 1
    return tmp
```

```python
import scipy.stats as stats                               ##   bandwidth 즉 h값을 조절하면 모양이 달라진다.
x1 = np.array([-7, -5, 1, 4, 5., 10, 13, 11, 12, 11.5, 18])
kde = stats.gaussian_kde(x1)
xs = np.linspace(-10, 20, num=100)
y1 = kde(xs)
kde.set_bandwidth(bw_method='silverman')
y2 = kde(xs)
kde.set_bandwidth(bw_method=kde.factor / 3.)
y3 = kde(xs)
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12,12))
ax.plot(x1, np.full(x1.shape, 1 / (4. * x1.size)), 'bo',
        label='Data points (rescaled)')
ax.plot(xs, y1, label='Scott (default)', linewidth=4, linestyle=':')
ax.plot(xs, y2, label='Silverman', linewidth=4, linestyle='--')
ax.plot(xs, y3, label='Const (1/3 * Silverman)', linewidth=4, linestyle='-')
ax.legend(fontsize=14)
ax.set_xlabel("X", fontsize=20)
ax.set_xticklabels(ax.get_xticks(),fontsize=18)
ax.set_ylabel("Density", fontsize=20)
ax.set_yticklabels(ax.get_yticks(),fontsize=18)
plt.savefig("kde_example.png", dpi=300)
```
